```python
def recomendar_cursos(intereses_alumno, semestre_actual, top_k=5):
    """
    Recomienda cursos según intereses del alumno y semestre.
    """

    resultados = collection.query(
        query_texts=[intereses_alumno],
        n_results=top_k,
        where={"semestre": {"$gte": semestre_actual}},
        include=['documents', 'metadatas', 'distances']
    )

    print("\n Recomendaciones:")
    print("=" * 50)

    for meta, dist in zip(resultados['metadatas'][0], resultados['distances'][0]):
        print(f"\n📘 {meta['nombre']}")
        print(f"Semestre: {meta['semestre']} | Área: {meta['area']}")
        print(f"Similitud: {1 - dist:.4f}")
```

```python
def cursos_similares(nombre_curso, n=3):
    """
    Encuentra cursos similares a uno dado.
    """

    # Buscar curso base
    datos = collection.get(where={"nombre": {"$eq": nombre_curso}})

    if not datos['documents']:
        print(" Curso no encontrado")
        return

    descripcion = datos['documents'][0]

    resultados = collection.query(
        query_texts=[descripcion],
        n_results=n + 1,
        include=['metadatas', 'distances']
    )

    print(f"\n Cursos similares a {nombre_curso}:")
    print("=" * 50)

    for meta, dist in zip(resultados['metadatas'][0][1:], resultados['distances'][0][1:]):
        print(f"{meta['nombre']} | Similitud: {1 - dist:.4f}")
```

```python
def buscar_por_perfil(habilidades_lista, area='especialidad'):
    """
    Busca cursos según habilidades del alumno.
    """

    consulta = " ".join(habilidades_lista)

    resultados = collection.query(
        query_texts=[consulta],
        n_results=5,
        where={"area": {"$eq": area}},
        include=['metadatas', 'distances']
    )

    print("\n Cursos según tu perfil:")
    print("=" * 50)

    for meta, dist in zip(resultados['metadatas'][0], resultados['distances'][0]):
        print(f"{meta['nombre']} | Similitud: {1 - dist:.4f}")
```

```python
def menu():
    while True:
        print("\n===== SISTEMA DE RECOMENDACIÓN =====")
        print("1. Recomendar cursos")
        print("2. Cursos similares")
        print("3. Buscar por perfil")
        print("4. Salir")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            intereses = input("Intereses: ")
            semestre = int(input("Semestre actual: "))
            recomendar_cursos(intereses, semestre)

        elif opcion == "2":
            nombre = input("Nombre del curso: ")
            cursos_similares(nombre)

        elif opcion == "3":
            habilidades = input("Habilidades (separadas por coma): ").split(",")
            buscar_por_perfil(habilidades)

        elif opcion == "4":
            print("👋 Saliendo...")
            break
```

```python
if __name__ == "__main__":
    menu()
```
